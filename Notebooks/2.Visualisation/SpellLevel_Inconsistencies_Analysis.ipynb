## Reorganising and renaming columns

```python

ReadCSV = True
FileName_IMD = "File_7_ID_2015_All_ranks__deciles_and_scores_for_the_Indices_of_Deprivation__and_population_denominators.csv"
df_IMD = OpenFile(ReadCSV, NewPath, FileName_IMD)


# Pre-process dataset: rename and reorganise features, convert date features
# --------------------------------------------------
def PreProcessing(df_ini, ConditionName, df_IMD):
    df = df_ini.copy()
    print('Pre-processing', ConditionName.upper(), 'dataset...')
    
    # Renaming features: ending in 1 (all three)
    FeaturesEnd1 = [x for x in df.columns if x[-1]=='1']
    for col in FeaturesEnd1:
        df = df.rename(columns={col:col[:-1]})
        
    # Renaming features: beginning with Right_ (diabetes)
    Features_Right = [x for x in df.columns if 'Right_' in x] 
    for col in Features_Right:
        df = df.rename(columns={col:col[6:]})

    # Converting date features
    DateFormat = '%Y-%m-%d'
    DateFeatures = ['Admission_date', 'Discharge_date', 'EPIstart', 'EPIend']
    for date in DateFeatures:
        df.loc[:, date] = pd.to_datetime( df.loc[:, date], format=DateFormat )  
    
    # Renaming existing frailty and deprivation data (autism)
    ColumnRename = ['IMD_score', 'IMD_decile', 'HFRS_score', 'HFRS_band']
    for col in ColumnRename:
        if col in df.columns:
            List = col.split('_')
            col_new = List[0] + '_' + List[1][0].upper()+List[1][1:]
            df = df.rename(columns={col:col_new})
            
    # Adding IMD data if missing (dementia and diabetes)
    if not('IMD_Score' in df.columns):
        ColumnList = ['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Score',
                      'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)']
        df_IMD_Sub = df_IMD.loc[:, ColumnList]

        df = pd.merge(df, df_IMD_Sub, left_on='LSOA_2011_Code', right_on='LSOA code (2011)', how='left')
        df = df.drop(columns=['LSOA code (2011)'])
        df = df.rename(columns={'Index of Multiple Deprivation (IMD) Score':'IMD_Score',
                                'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)':'IMD_Decile'})

    # for DIABETES: Keep ethnicity Category
    if 'Category' in df.columns:
        df = df.drop(columns=['Ethnicity_Description']).rename(columns={'Category':'Ethnicity'})

    # Reorganising columns
    ColumnList = ['P_Spell_ID', 'HESID', 'LSOA_2011_Code',
                  'Admission_date', 'Discharge_date', 'EPIstart', 'EPIend',
                  'diagnosis_group', 'procedure_group',
                  'Spell_Los', 'sex', 'age_of_patient', 'Ethnicity',
                  'IMD_Score', 'IMD_Decile', 'HFRS_Score', 'HFRS_Band',
                  'ProvCode', 'Main_Specialty_Description', 'POD',
                  'MORT', 'Read30', 'Read90']
    
    MissingColumn = [x for x in ColumnList if not(x in df.columns)]
    if len(MissingColumn)>0:
        print('Missing Feature:', MissingColumn)

    df = df.loc[:, ColumnList]
    
    return df
# --------------------------------------------------

df_Autism_Processed = PreProcessing(df_Autism, 'Autism', df_IMD)
df_Diabetes_Processed = PreProcessing(df_Diabetes, 'Diabetes', df_IMD)
df_Dementia_Processed = PreProcessing(df_Dementia, 'Dementia', df_IMD)

SummaryDataFrame(df_Autism_Processed, 'Autism')
SummaryDataFrame(df_Diabetes_Processed, 'Diabetes')
SummaryDataFrame(df_Dementia_Processed, 'Dementia')
```

## Find ICD-10 chains at spell level

```python


# Converting List of ICD-10 codes to string without duplicates
# --------------------------------------------------
def ICD10Chain_ToList(List):
    List2 = list( dict.fromkeys(List) )
    List2 = [x for x in List2 if x!='-1']
    String = '~'.join(List2)

    return String
# --------------------------------------------------

# Add SpellLevel_ICD10 feature to dataframe and keep last episode of each spell
# --------------------------------------------------
def SpellLevel_ICD10_OPCS(df_ini):
    df = df_ini.copy()

    # Find ICD-10 codes across spells
    ICD10_Spells = df.groupby(by=['P_Spell_ID'])['diagnosis_group'].agg('~'.join).reset_index()
    ICD10_Spells.loc[:, 'diagnosis_group'] = ICD10_Spells.loc[:, 'diagnosis_group'].apply(lambda x: ICD10Chain_ToList( x.split('~') ) )

    df_Merged = pd.merge(df, ICD10_Spells, on='P_Spell_ID', how='left').rename(columns={'diagnosis_group_x':'diagnosis_group',
                                                                                        'diagnosis_group_y':'SpellLevel_ICD10'})
    
    # Find OPCS codes across spells
    OPCS_Spells = df.groupby(by=['P_Spell_ID'])['procedure_group'].agg('~'.join).reset_index()
    OPCS_Spells.loc[:, 'procedure_group'] = OPCS_Spells.loc[:, 'procedure_group'].apply(lambda x: ICD10Chain_ToList( x.split('~') ) )
    
    df_Merged = pd.merge(df_Merged, OPCS_Spells, on='P_Spell_ID', how='left').rename(columns={'procedure_group_x':'procedure_group',
                                                                                              'procedure_group_y':'SpellLevel_OPCS'})

    # Keep only last episode of each spell
    df_Merged = df_Merged.drop_duplicates(subset=['P_Spell_ID'], keep='last')
    
    return df_Merged
# --------------------------------------------------

df_Autism_SpellLevel = SpellLevel_ICD10_OPCS(df_Autism_Processed)
df_Diabetes_SpellLevel = SpellLevel_ICD10_OPCS(df_Diabetes_Processed)
df_Dementia_SpellLevel = SpellLevel_ICD10_OPCS(df_Dementia_Processed)

SummaryDataFrame(df_Autism_SpellLevel, 'Autism')
SummaryDataFrame(df_Diabetes_SpellLevel, 'Diabetes')
SummaryDataFrame(df_Dementia_SpellLevel, 'Dementia')
```

## Finding mandatory codes before identifying inconsistencies

```python

# Find mentions of mandatory codes in CodeList in ICDfeature_Name
# --------------------------------------------------
def SpellLevel_FindMandatoryCode(df_ini, ICDfeature_Name, ConditionName, CodeList):
    df = df_ini
    
    Codes = '|'.join(CodeList)

    # Look for ANY mention of mandatory code
    Filter = df[ICDfeature_Name].str.contains(Codes)
    df.loc[:, "SpellLevel_ICD10_" + ConditionName] = np.where(Filter, 1, 0)

    return df
# --------------------------------------------------

# Mandatory codes for dementia
CodesDementia_Categories = "F00|F01|F02"
CodesDementia_Unspecified = "F03|F051"
Codes_Dementia_SubCategories = "F000|F001|F002|F009|F010|F011|F012|F013|F018|F019|F020|F021|F022|F023|F024|F028"
Codes_Dementia_Supp = "G300|G301|G308|G309" # No G302, G300 instead!
CodesDementia = CodesDementia_Categories + "|" + CodesDementia_Unspecified + "|" + Codes_Dementia_SubCategories + "|" + Codes_Dementia_Supp
CodesDementia_List = CodesDementia.split("|")
Parkinson_List = ["G20"] 
ParkinsonsDementia_List = ["F023"]

# Mandatory codes for autism
Codes_LearningDisabilities = ['F7' + i for i in ['0', '1', '2', '3', '8', '9']]
CodesAutism_List = ["F840", "F841", "F845"]

# Mandatory codes for diabetes
CodesDiabetes_List = ["E10", "E11", "E14"]
CodesDiabetesTypeII_List = ["E11"]

# Look for mandatory codes
df_Autism_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_Autism_SpellLevel, "SpellLevel_ICD10", "Autism", CodesAutism_List)
df_Autism_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_Autism_SpellLevel_2, "SpellLevel_ICD10", "LearningDisabilities", Codes_LearningDisabilities)

df_Diabetes_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_Diabetes_SpellLevel, "SpellLevel_ICD10", "Diabetes", CodesDiabetes_List)
df_Diabetes_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_Diabetes_SpellLevel_2, "SpellLevel_ICD10", "DiabetesTypeII", CodesDiabetesTypeII_List)

df_Dementia_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_Dementia_SpellLevel, "SpellLevel_ICD10", "Dementia", CodesDementia_List)
df_Dementia_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_Dementia_SpellLevel_2, "SpellLevel_ICD10", "Parkinson", Parkinson_List)
df_Dementia_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_Dementia_SpellLevel_2, "SpellLevel_ICD10", "ParkinsonsDementia", ParkinsonsDementia_List)


SummaryDataFrame(df_Autism_SpellLevel_2, 'Autism')
SummaryDataFrame(df_Diabetes_SpellLevel_2, 'Diabetes')
SummaryDataFrame(df_Dementia_SpellLevel_2, 'Dementia')

```

```python

# --------------------------------------------------
def ErrorCharacteristics_SpellLevel(df_ini, ConditionName):
    df = df_ini.copy()
    print('-----------------------------------------------------')
    print(BoldFont(ConditionName.upper()), ':')
    
    Filter_SubsequentEpisodes = (df.loc[:, 'SubsequentSpell']==1)
    df = df.loc[Filter_SubsequentEpisodes, :]
    
    ErrorName = 'SpellLevel_Error_'+ConditionName
    Filter = (df.loc[:, ErrorName]==1)
    df_Errors = df.loc[Filter, :]
    df_NoErrors = df.loc[~Filter, :]
    
    print('Total number of subsequent spell:', df.shape[0])
    print('Number of inconsistencies at spell level:', df_Errors.shape[0])
    print('Proportion of coding inconsistencies: %0.2f' % (100*df_Errors.shape[0]/df.shape[0]), '%')
    
    
    # --------------------------------------------
    # Errors overnight stays
    Filter_0LOS = df.loc[:, 'Spell_Los']==0
    df_0LOS = df[Filter_0LOS]
    df_Not0LOS = df[~Filter_0LOS]
    def Errors_Print(df_Sub):
        N_Errors = df_Sub.loc[:, ErrorName].sum()
        N_SubsequentSpells = df_Sub.loc[:, 'SubsequentSpell'].sum()
        Prop_Errors = 100 * N_Errors / N_SubsequentSpells
        print('Number of subsequent spells:', f"{N_SubsequentSpells:,}")
        print('Number of inconsistencies:', f"{N_Errors:,}",
              'corresponding to %0.2f' % Prop_Errors, '%')
        
    print(BoldFont('Overnight stays:'))
    print('WITH overnight stay:')
    Errors_Print(df_Not0LOS)
    print('Day-case:')
    Errors_Print(df_0LOS)
    print('\n')
    
    # --------------------------------------------
    # Errors by method of admission
    Filter_PODNan = df.loc[:, 'POD'].isnull()
    Filter_EM = df.loc[~Filter_PODNan, 'POD'].str.contains('EM|NE')
    Filter_EL = df.loc[~Filter_PODNan, 'POD'].str.contains('EL|DC|RA')
    df_EM = df[(~Filter_PODNan)&Filter_EM]
    df_EL = df[(~Filter_PODNan)&Filter_EL]
    df_PODNan = df[Filter_PODNan]

    print('POD, Emergency:')
    Errors_Print(df_EM)     
    print('POD, Elective:')
    Errors_Print(df_EL)     
    print('POD, Not recorded:')
    Errors_Print(df_PODNan)     
    print('\n')
    
    # --------------------------------------------
    # Errors by age bands
    AgeMin_List = [0, 18, 40, 60, 80] + [999]
    AgeMax_List = [age-1 for age in AgeMin_List[1:] ] + [999]
    for k in range(len(AgeMin_List)):
        AgeMin = AgeMin_List[k]
        AgeMax = AgeMax_List[k]
        FilterAge = (df['age_of_patient']>=AgeMin) & (df['age_of_patient']<=AgeMax)

        df_Age = df.loc[FilterAge, :]
        Sum = df_Age.loc[:, ErrorName].sum()
        Count = df_Age.loc[:, ErrorName].count()
        Prop_Errors = 100 * Sum/Count
        print(BoldFont('Age band:'), AgeMin, '-', AgeMax)
        print('\t Number of inconsistencies:', f"{Sum:,}", 'out of', f"{Count:,}", ', corresponding to %0.2f' % Prop_Errors, '%')

    print('\n')

    # --------------------------------------------
    # Errors by sex
    List = ['Female', 'Male']
    for sex in List:
        Filter_sex = (df.loc[:, 'sex']==sex)
        df_sex =  df.loc[Filter_sex, :]
        Sum = df_sex.loc[:, ErrorName].sum()
        Count = df_sex.loc[:, ErrorName].count()
        Prop_Errors = 100*Sum/Count
        print(BoldFont('Sex:'), sex)
        print('\t Number of inconsistencies:', f"{Sum:,}", 'out of', f"{Count:,}", ', corresponding to %0.2f' % Prop_Errors, '%')

    SexString = '|'.join(List)
    Filter_Sex = df.loc[:, 'sex'].str.contains(SexString)
    df_MissingSex = df.loc[~Filter_Sex, :]
    Sum = df_MissingSex.loc[:, ErrorName].sum()
    Count = df_MissingSex.loc[:, ErrorName].count()
    Prop_Errors = 100*Sum/Count
    print('Sex:', 'Other/Missing')
    print('\t Number of inconsistencies:', f"{Sum:,}", 'out of', f"{Count:,}", ', corresponding to %0.2f' % Prop_Errors, '%')
    print('\n')
    
    print('\n')
    print(BoldFont('Ethnicity'))
    # --------------------------------------------
    # Ethnicity
    if 'Ethnicity' in df.columns:
        df.loc[:, 'Ethnicity'] = df.loc[:, 'Ethnicity'].fillna('not known')
        df.loc[:, 'Ethnicity'] = df.loc[:, 'Ethnicity'].str.lower()
        EthnicityList = [ x for x in set(df.loc[:, 'Ethnicity'].values) ]
#         Count = df.shape[0]
        for ethnicity in EthnicityList:
            Filter = df.loc[:, 'Ethnicity']==ethnicity
            Sum = df.loc[Filter, ErrorName].sum()
            Count = df.loc[Filter, ErrorName].shape[0] 
            Prop = 100*Sum/Count
            print('Number of', ethnicity, ' :', f"{Sum:,}", 'corresponding to %0.2f' % Prop, '%')

    print('\n')
    # Errors by IMD deciles
    # --------------------------------------------
#     # Deprivation Quintile
#     ColumnList = ['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)']
#     df_IMD_Sub = df_IMD[ColumnList]

#     df = pd.merge(df, df_IMD_Sub, left_on='LSOA_2011_Code', right_on='LSOA code (2011)', how='left')
#     df = df.drop(columns=['LSOA code (2011)'])
#     df = df.rename(columns={'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)':'IMD Deciles'})

    print(BoldFont('IMD Score'))
    if 'IMD_Decile' in df.columns:
        Decile_List = range(1,10+1)
        for k in range(5):
            Decile_Max = Decile_List[k]*2
            Decile_Min = Decile_Max-1

            Filter_Quintiles = (df["IMD_Decile"]>=Decile_Min) & (df["IMD_Decile"]<=Decile_Max)
            Sum = df.loc[Filter_Quintiles, ErrorName].sum()
            Count = df.loc[Filter_Quintiles, ErrorName].count()
            Prop_Errors = 100*Sum/Count
            print('IMD Deciles:', Decile_Min, '-', Decile_Max)
            print('\t Number of errors:', f"{Sum:,}", 'out of', f"{Count:,}", ', corresponding to %0.2f' % Prop_Errors, '%')
        Filter_IMDNaN = df["IMD_Decile"].isnull()
        df_IMDNaN = df[Filter_IMDNaN]
        print('IMD not recorded')
        Errors_Print(df_IMDNaN)
# --------------------------------------------------

ErrorCharacteristics_SpellLevel(df_Autism_SpellLevel_3, 'Autism')
# ErrorCharacteristics_SpellLevel(df_Diabetes_SpellLevel_3, 'Diabetes')
# ErrorCharacteristics_SpellLevel(df_Dementia_SpellLevel_3, 'Dementia')

```

```python

# --------------------------------------------------
def TimeEvolution_Quarters(df_ini, ConditionName):
    DateMin = df_Dementia_SpellLevel_3['FirstSpell_Discharge_date'].min()
    DateMax = df_Dementia_SpellLevel_3['FirstSpell_Discharge_date'].max()
    ThreeYr_Days = 3*365.25

    # Initialisation
    DateList_FirstSpell = pd.date_range(start=DateMin, end=DateMax)
    N_SubsequentSpells = []
    N_Inconsistencies = []
    Prop_Inconsistencies = []

    # Loop through discharge date of first spells
    for date in DateList_FirstSpell:
        Filter_FirstSpellDate = (df_ini['FirstSpell_Discharge_date']==date)
        Filter_ThreeYr = (df_ini['TimeDiff_SinceFirst']<=ThreeYr_Days)
        Filter_Both = Filter_FirstSpellDate & Filter_ThreeYr
        df_Sub = df_ini.loc[Filter_Both, :]

        n_Inconsistencies = df_Sub.loc[:, 'SpellLevel_Error_'+ConditionName].sum()
        n_SubsequentSpells = df_Sub.loc[:, 'SubsequentSpell'].sum()

        N_Inconsistencies.append(n_Inconsistencies)
        N_SubsequentSpells.append(n_SubsequentSpells)

    df_TimeEvolution = pd.DataFrame(data={'FirstSpell_DischargeDate':DateList_FirstSpell,
                                          'Number_SubsequentSpells':N_SubsequentSpells,
                                          'Number_Inconsistencies':N_Inconsistencies})

    # Grouping by %Y-%m discharge dates of first spells
    # df_TimeEvolution.loc[:, 'FirstSpell_DischargeDate_YrMonth'] = df_TimeEvolution['FirstSpell_DischargeDate'].dt.strftime('%Y-%m')

    # df_TimeEvolution.loc[:, 'Number_SubsequentSpells_YrMonth'] = df_TimeEvolution.groupby('FirstSpell_DischargeDate_YrMonth')['Number_SubsequentSpells'].transform('sum')
    # df_TimeEvolution.loc[:, 'Number_Inconsistencies_YrMonth'] = df_TimeEvolution.groupby('FirstSpell_DischargeDate_YrMonth')['Number_Inconsistencies'].transform('sum')
    # df_TimeEvolution.loc[:, 'Prop_Inconsistencies_YrMonth'] = 100 * df_TimeEvolution.loc[:, 'Number_Inconsistencies_YrMonth'] / df_TimeEvolution.loc[:, 'Number_SubsequentSpells_YrMonth']
    # ColumnList = ['FirstSpell_DischargeDate_YrMonth', 'Number_Inconsistencies_YrMonth',
    #               'Number_SubsequentSpells_YrMonth', 'Prop_Inconsistencies_YrMonth']
    # df_TimeEvolution_YrMonth = df_TimeEvolution.drop_duplicates(subset=['FirstSpell_DischargeDate_YrMonth'])[ColumnList]
    # df_TimeEvolution_YrMonth.head(10)

    df = df_TimeEvolution
    #  Group in quarters
    df_Quarters = df.groupby(df['FirstSpell_DischargeDate'].dt.to_period('Q'))['Number_Inconsistencies'].agg('sum').reset_index()
    df_Quarters.loc[:, 'Number_SubsequentSpells'] = df.groupby(df['FirstSpell_DischargeDate'].dt.to_period('Q'))['Number_SubsequentSpells'].agg('sum').reset_index()

    df_Quarters.loc[:, 'Prop_Inconsistencies'] = 100 * df_Quarters.loc[:, 'Number_Inconsistencies'] / df_Quarters.loc[:, 'Number_SubsequentSpells']
    df_Quarters = df_Quarters.iloc[:-1, :]
    
    return df_Quarters
# --------------------------------------------------

df_Quarters_Autism = TimeEvolution_Quarters(df_Autism_SpellLevel_3, 'Autism')
df_Quarters_Diabetes = TimeEvolution_Quarters(df_Diabetes_SpellLevel_3, 'Diabetes')
df_Quarters_Dementia = TimeEvolution_Quarters(df_Dementia_SpellLevel_3, 'Dementia')

```
