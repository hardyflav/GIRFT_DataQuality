### Reorganising and renaming columns

```python

print('Importing IMD data...')
Path_IMD = r"O:\GIRFT DiRAC\FlavienHardy\1_DataConsistency\Data\0_Raw"
FileName_IMD = "File_7_ID_2015_All_ranks__deciles_and_scores_for_the_Indices_of_Deprivation__and_population_denominators.csv"
FilePath_IMD = os.path.join(Path_IMD, FileName_IMD)
df_IMD = OpenFile(Path_IMD, FilePath_IMD)
        

# Pre-process dataset: rename and reorganise features, convert date features
# --------------------------------------------------
def PreProcessing(df_ini, ConditionName, df_IMD):
    df = df_ini.copy()
    print('Pre-processing', ConditionName.upper(), 'dataset...')
    
    # Renaming features: ending in 1 (all three)
    FeaturesEnd1 = [x for x in df.columns if x[-1]=='1']
    for col in FeaturesEnd1:
        df = df.rename(columns={col:col[:-1]})
        
    # Renaming features: beginning with Right_ (diabetes)
    Features_Right = [x for x in df.columns if 'Right_' in x] 
    for col in Features_Right:
        df = df.rename(columns={col:col[6:]})

    # Converting date features
    DateFormat = '%Y-%m-%d'
    DateFeatures = ['Admission_date', 'Discharge_date', 'EPIstart', 'EPIend']
    for date in DateFeatures:
        df.loc[:, date] = pd.to_datetime( df.loc[:, date], format=DateFormat )  
    
    # Renaming existing frailty and deprivation data (autism)
    ColumnRename = ['IMD_score', 'IMD_decile', 'HFRS_score', 'HFRS_band']
    for col in ColumnRename:
        if col in df.columns:
            List = col.split('_')
            col_new = List[0] + '_' + List[1][0].upper()+List[1][1:]
            df = df.rename(columns={col:col_new})
            
    # Adding IMD data if missing (dementia and diabetes)
    if not('IMD_Score' in df.columns):
        ColumnList = ['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Score',
                      'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)']
        df_IMD_Sub = df_IMD.loc[:, ColumnList]

        df = pd.merge(df, df_IMD_Sub, left_on='LSOA_2011_Code', right_on='LSOA code (2011)', how='left')
        df = df.drop(columns=['LSOA code (2011)'])
        df = df.rename(columns={'Index of Multiple Deprivation (IMD) Score':'IMD_Score',
                                'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)':'IMD_Decile'})

    # for DIABETES: Keep ethnicity Category
    if 'Category' in df.columns:
        df = df.drop(columns=['Ethnicity_Description']).rename(columns={'Category':'Ethnicity'})

    # Reorganising columns
    ColumnList = ['P_Spell_ID', 'HESID', 'LSOA_2011_Code',
                  'Admission_date', 'Discharge_date', 'EPIstart', 'EPIend',
                  'diagnosis_group', 'procedure_group',
                  'Spell_Los', 'sex', 'age_of_patient', 'Ethnicity',
                  'IMD_Score', 'IMD_Decile', 'HFRS_Score', 'HFRS_Band',
                  'ProvCode', 'Main_Specialty_Description', 'POD',
                  'MORT', 'Read30', 'Read90']
    
    MissingColumn = [x for x in ColumnList if not(x in df.columns)]
    if len(MissingColumn)>0:
        print('Missing Feature:', MissingColumn)

    df = df.loc[:, ColumnList]
    
    return df
# --------------------------------------------------

df_dementia = PreProcessing(df_dementia_ini, 'Dementia', df_IMD)
df_autism = PreProcessing(df_autism_ini, 'Autism', df_IMD)
df_diabetes = PreProcessing(df_diabetes_ini, 'Diabetes', df_IMD)

SummaryDataFrame(df_dementia, 'Dementia')
SummaryDataFrame(df_autism, 'Autism')
SummaryDataFrame(df_diabetes, 'Diabetes')
```

### Identify missing mandatory codes

```python

# Find mentions of mandatory codes in CodeList in ICDfeature_Name
# --------------------------------------------------
def FindMandatoryCode(df_ini, ICDfeature_Name, ConditionName, CodeList):
    df = df_ini
    
    Codes = '|'.join(CodeList)

    # Look for ANY mention of mandatory code
    Filter = df[ICDfeature_Name].str.contains(Codes)
    df.loc[:, "ICD10_" + ConditionName] = np.where(Filter, 1, 0)

    return df
# --------------------------------------------------

# Mandatory codes for dementia
CodesDementia_Categories = "F00|F01|F02"
CodesDementia_Unspecified = "F03|F051"
Codes_Dementia_SubCategories = "F000|F001|F002|F009|F010|F011|F012|F013|F018|F019|F020|F021|F022|F023|F024|F028"
Codes_Dementia_Supp = "G300|G301|G308|G309" # No G302, G300 instead!
CodesDementia = CodesDementia_Categories + "|" + CodesDementia_Unspecified + "|" + Codes_Dementia_SubCategories + "|" + Codes_Dementia_Supp
CodesDementia_List = CodesDementia.split("|")
Parkinson_List = ["G20"] 
ParkinsonsDementia_List = ["F023"]

# Mandatory codes for autism
CodesAutism_List = ["F840", "F841", "F845"]

# Mandatory codes for diabetes
CodesDiabetes_List = ["E10", "E11", "E14"]
CodesDiabetesTypeII_List = ["E11"]


# Look for mandatory codes
df_dementia_2 = FindMandatoryCode(df_dementia, "diagnosis_group", "Dementia", CodesDementia_List)
df_dementia_2 = FindMandatoryCode(df_dementia_2, "diagnosis_group", "Parkinson", Parkinson_List)
df_dementia_2 = FindMandatoryCode(df_dementia_2, "diagnosis_group", "ParkinsonsDementia", ParkinsonsDementia_List)

df_autism_2 = FindMandatoryCode(df_autism, "diagnosis_group", "Autism", CodesAutism_List)

df_diabetes_2 = FindMandatoryCode(df_diabetes, "diagnosis_group", "Diabetes", CodesDiabetes_List)
df_diabetes_2 = FindMandatoryCode(df_diabetes_2, "diagnosis_group", "DiabetesTypeII", CodesDiabetesTypeII_List)


SummaryDataFrame(df_dementia_2, 'Dementia')
SummaryDataFrame(df_autism_2, 'Autism')
SummaryDataFrame(df_diabetes_2, 'Diabetes')

```

```python
# --------------------------------------------------
def FindErrors(df_ini, ConditionName):
    df = df_ini
    print(ConditionName.upper())
    
    # Exclude all spells with discharge date > end cut-off-date
    Filter_NotEnd = (df.loc[:, 'Discharge_date']<=pd.to_datetime('2021-03-31', format='%Y-%m-%d'))
    df = df.loc[Filter_NotEnd, :]

    # Exclude all spells with discharge date < start cut-off-date
    Filter_AfterFirstTwoYears = (df.loc[:, 'Discharge_date']>=pd.to_datetime('2013-04-01', format='%Y-%m-%d'))
    df = df.loc[Filter_AfterFirstTwoYears, :]
    
# ---------------- Removing particular spells    
#     Exclude specific age bands and other conditions; only AFTER summarising results in Table 2.1
    if ConditionName=='Dementia':
#         Remove <40 year-old patients
        Filter_Age = df.loc[:, 'age_of_patient']>=40
        print('Number of episodes removed because age < 40:', df[~Filter_Age].shape[0])
        df = df.loc[Filter_Age, :]    
    if ConditionName=='Diabetes':
#         Remove <18 year-old patients
        Filter_Age = df.loc[:, 'age_of_patient']>=18
        print('Number of episodes removed because age < 18:', df[~Filter_Age].shape[0])
        df = df.loc[Filter_Age, :]    
        # Dialysis patients
        Filter_OPCSDialysis = df.loc[:, 'procedure_group'].str.contains('X40|X43')
        df.loc[:, 'ICD10_Dialysis'] = np.where(Filter_OPCSDialysis, 1, 0)
        Filter_NoDialysis =  df.loc[:, 'ICD10_Dialysis']==0
        print('Number of episodes removed because of dialysis treatment:', df.loc[:, 'ICD10_Dialysis'].sum(),
          'corresponding to %0.2f' % (100*Filter_OPCSDialysis.sum()/Filter_OPCSDialysis.shape[0]), '%')
        df = df.loc[Filter_NoDialysis, :]
        
    # Remove explicit errors: ICD-10 chains with only R69 or N185 codes
    def ProcessList(L):
        L_Processed = [x for x in L if x!='-1']
        L_String = '~'.join(L_Processed)
        return L_String

    df.loc[:, 'diagnosis_group_Processed'] = df.loc[:, 'diagnosis_group'].str.split('~').apply(lambda x: ProcessList(x))
    Filter_R69 = (df.loc[:, 'diagnosis_group_Processed'].str.contains('R69'))
    Filter_MoreThanOneCode =  df.loc[:, 'diagnosis_group_Processed'].str.contains('~')
    Filter_OnlyR69 = Filter_R69 & (~Filter_MoreThanOneCode)
    df = df.loc[~Filter_OnlyR69, :]
    print('Numbers of episodes with only R69 codes:', Filter_OnlyR69.sum(),
          'corresponding to %0.2f' % (100*Filter_OnlyR69.sum()/Filter_OnlyR69.shape[0]), '%')
    
    if ConditionName=='Diabetes':
        Filter_N185 = (df.loc[:, 'diagnosis_group_Processed'].str.contains('N185'))
        Filter_MoreThanOneCode =  df.loc[:, 'diagnosis_group_Processed'].str.contains('~')
        Filter_0LOS = df.loc[:, 'Spell_Los']==0
        Filter_Only185 = Filter_0LOS & Filter_N185 & (~Filter_MoreThanOneCode)
        
        df = df.loc[~Filter_Only185, :].drop(columns=['diagnosis_group_Processed'])
        print('Numbers of episodes with only N185 codes:', Filter_Only185.sum(),
          'corresponding to %0.2f' % (100*Filter_Only185.sum()/Filter_Only185.shape[0]), '%')
        print('Episodes removed.')
# ---------------- Removing particular spells    

    Filter_Code = df["ICD10_"+ConditionName] == 1
    df_Code = df.loc[Filter_Code, :]
    df_NoCode = df.loc[~Filter_Code, :]

    # Keep first occurence of mandatory code
    df_CodeFirst = df_Code.sort_values(by=['HESID', 'EPIstart', 'EPIend']).drop_duplicates(subset=['HESID'], keep='first')

    ColumnList = ['HESID', 'P_Spell_ID',
                  'EPIstart', 'EPIend', 'Admission_date', 'Discharge_date',
                  'ProvCode', 'POD', 'Main_Specialty_Description']
    
    df = pd.merge( df, df_CodeFirst[ColumnList], how='left', on='HESID' )
    
    ColumnRename_Original = [x for x in df.columns if '_x' in x]
    ColumnRename_First = [x for x in df.columns if '_y' in x]
    ColumnRename = ColumnRename_Original + ColumnRename_First
    
    d_rename = dict(zip(ColumnRename, ColumnRename))
    for feature in ColumnRename:
        if '_x' in feature:
            d_rename[feature] = feature[:-2]
        elif 'y' in feature:
            d_rename[feature] = 'FirstEpisode_' + feature[:-2]
    
    df = df.rename(columns=d_rename)
    
    # Flag first episode
    df_CodeFirst.loc[:, 'FirstDiagnosis'] = 1
    ColumnList = ['P_Spell_ID', 'EPIstart', 'EPIend', 'FirstDiagnosis']
    df = pd.merge(df, df_CodeFirst[ColumnList], on=['P_Spell_ID', 'EPIstart', 'EPIend'], how='left')
    df.loc[:, 'FirstDiagnosis'] = df.loc[:, 'FirstDiagnosis'].fillna(0)
    df = df.astype({'FirstDiagnosis':int})
    
#     FirstDiagnosis_idx = df.index.isin(df_CodeFirst.index)
#     df.loc[:, 'FirstDiagnosis']=np.where(FirstDiagnosis_idx, 1, 0)
    
    # Flag subsequent episodes
    Filter_SubsequentEpisode = (df['FirstEpisode_EPIend']<=df['EPIend']) & (df.loc[:, 'FirstDiagnosis']==0)
    df.loc[:, 'SubsequentEpisode'] = np.where(Filter_SubsequentEpisode, 1, 0)
    
    # Find errors in dataset
    Filter_Error = (df["ICD10_"+ConditionName]==0) & (df['FirstEpisode_EPIend']<df['EPIend'])
    df.loc[:, 'Error_'+ConditionName] = np.where(Filter_Error, 1, 0)

    df = df.sort_values(by=['HESID', 'EPIstart', 'EPIend'])
    
    return df
# --------------------------------------------------

df_dementia_3 = FindErrors(df_dementia_2, "Dementia")
df_autism_3 = FindErrors(df_autism_2, "Autism")
df_diabetes_3 = FindErrors(df_diabetes_2, "Diabetes")

SummaryDataFrame(df_dementia_3, 'Dementia')
SummaryDataFrame(df_autism_3, 'Autism')
SummaryDataFrame(df_diabetes_3, 'Diabetes')

# displayDataFrame(df_dementia_3.head(2))
# displayDataFrame(df_autism_3.head(2))
# displayDataFrame(df_diabetes_3.head(2))
```

### Count coding errors

```python

# Count number of errors
# --------------------------------------------------
def CountErrors(df_ini, NameCondition):
    df = df_ini.copy()
    
    # Keep episodes with existing dementia diagnosis (e.g. no G20X only episodes)
    Filter_NotFirst = (df['SubsequentEpisode']==1)
    df = df.loc[Filter_NotFirst, :]

    Sum = df['Error_'+NameCondition].sum()
    Count = df['Error_'+NameCondition].shape[0]
    Prop = 100 * Sum / Count
    
    return Sum, Count, Prop

# --------------------------------------------------

Sum_dementia, Count_dementia, Prop_dementia = CountErrors(df_dementia_3, 'Dementia')
print('-----------------------------')
print('DEMENTIA')
print('Number of errors:', Sum_dementia)
print('Number of subsequent episodes where mandatory codes should be used:', Count_dementia)
print('Proportion of errors: %0.2f' % Prop_dementia, '%')
Filter_SubsEpisodes = (df_dementia_3['SubsequentEpisode']==1)
Sum_G20 = df_dementia_3.loc[Filter_SubsEpisodes, 'ICD10_Parkinson'].sum()
Count_G20 = df_dementia_3.loc[Filter_SubsEpisodes, 'ICD10_Parkinson'].shape[0]
Prop_G20 = 100 * (Count_G20-Sum_G20) / Count_G20
print('Proportion of subsequent episodes with no mention of PARKINSON: %0.2f' % Prop_G20, '%')

Sum_autism, Count_autism, Prop_autism = CountErrors(df_autism_3, 'Autism')
print('-----------------------------')
print('AUTISM')
print('Number of errors:', Sum_autism)
print('Number of subsequent episodes where mandatory codes should be used:', Count_autism)
print('Proportion of errors: %0.2f' % Prop_autism, '%')

Sum_diabetes, Count_diabetes, Prop_diabetes = CountErrors(df_diabetes_3, 'Diabetes')
print('-----------------------------')
print('DIABETES')
print('Number of errors:', Sum_diabetes)
print('Number of subsequent episodes where mandatory codes should be used:', Count_diabetes)
print('Proportion of errors: %0.2f' % Prop_diabetes, '%')
```

```python

# Count error rate in spells with no overnight stay
def CountErrors_0LOS(df_ini, ConditionName):
    df = df_ini.copy()
    df.loc[:, 'Episode_Los'] = (df.loc[:, 'EPIend']-df.loc[:, 'EPIstart']).dt.days
    Filter_0LOS = (df.loc[:, 'Episode_Los']==0)
    df_0LOS = df[Filter_0LOS]
    df_LOS = df[~Filter_0LOS]
    Sum_0LOS, Count_0LOS, Prop_0LOS = CountErrors(df_0LOS, ConditionName)
    Sum_LOS, Count_LOS, Prop_LOS = CountErrors(df_LOS, ConditionName)
    print('-----------------------------')
    print(ConditionName.upper() + ' - > 0 LOS')
    print('Number of errors:', Sum_LOS)
    print('Number of subsequent episodes where mandatory codes should be used:', Count_LOS)
    print('Proportion of errors: %0.2f' % Prop_LOS, '%')
    print(ConditionName.upper() + ' - 0 LOS')
    print('Number of errors:', Sum_0LOS)
    print('Number of subsequent episodes where mandatory codes should be used:', Count_0LOS)
    print('Proportion of errors: %0.2f' % Prop_0LOS, '%')

CountErrors_0LOS(df_autism_3, 'Autism')
CountErrors_0LOS(df_diabetes_3, 'Diabetes')
CountErrors_0LOS(df_dementia_3, 'Dementia')
```

```python

def ProfileFirstSpells(df_ini):

    # Find Unique First Spells
    df_First = df_ini.sort_values(by=['HESID', 'Admission_date', 'EPIstart', 'EPIend']).drop_duplicates(subset=['HESID'], keep='first')
#     print(df_First.loc[:, 'SpellLevel_TimeDiff'].describe())
    
    # --------------------------------------------
    df_Table1 = df_First # Unique patients, first spells
    N_Patients = df_Table1.shape[0]
    print('Number of unique patients/first spells:', N_Patients)
    print('\n')

    AgeMin_List = [0, 18, 40, 60, 80] + [999]
    AgeMax_List = [age-1 for age in AgeMin_List[1:] ] + [999]
    Filter_AgeRecorded = df_Table1['age_of_patient']<999
    N_AgeRecorded = df_Table1.loc[Filter_AgeRecorded, 'age_of_patient'].shape[0]
    for k in range(len(AgeMin_List)):
        AgeMin = AgeMin_List[k]
        AgeMax = AgeMax_List[k]
        FilterAge = (df_Table1['age_of_patient']>=AgeMin) & (df_Table1['age_of_patient']<=AgeMax)

        df_Age = df_Table1.loc[FilterAge, :]
        print('Age band:', AgeMin, '-', AgeMax)
        print('\t Number of patients:', df_Age.shape[0], 'corresponding to %0.2f' % (100*df_Age.shape[0]/N_AgeRecorded), '%')

    print('\n')

    # --------------------------------------------
    # Sex
    Filter_Male = df_Table1['sex']=='Male'
    Filter_Female = df_Table1['sex']=='Female'

    N_Males = df_Table1[Filter_Male].shape[0]
    N_Females = df_Table1[Filter_Female].shape[0]
    Missing_Sex = N_Patients - (N_Males + N_Females)
    Prop_Males = 100*N_Males/(N_Males + N_Females)
    Prop_Females = 100*N_Females/(N_Males + N_Females)
    Prop_Missing = 100*Missing_Sex/(N_Males + N_Females)

    print('Number of Male patients:', N_Males, 'corresponding to %0.2f' % Prop_Males, '%' )
    print('Number of Female patients:', N_Females, 'corresponding to %0.2f' % Prop_Females, '%' )
    print('Number of missing sex:', Missing_Sex, 'corresponding to %0.2f' % Prop_Missing, '%' )

    print('\n')
    # --------------------------------------------
    # Ethnicity
    if 'Ethnicity' in df_Table1.columns:
        df_Table1.loc[:, 'Ethnicity'] = df_Table1.loc[:, 'Ethnicity'].fillna('not known')
        df_Table1.loc[:, 'Ethnicity'] = df_Table1.loc[:, 'Ethnicity'].str.lower()

        EthnicityList = [ x for x in set(df_Table1.loc[:, 'Ethnicity'].values) ]
        Count = df_Table1.shape[0]
        Filter_NotKnown = (df_Table1.loc[:, 'Ethnicity']=='not known')
        N_NotKnown = df_Table1.loc[Filter_NotKnown, 'Ethnicity'].shape[0]
        for ethnicity in EthnicityList:
            Filter = df_Table1.loc[:, 'Ethnicity']==ethnicity
            Sum = df_Table1.loc[Filter, :].shape[0]
            Prop = 100*Sum/(Count-N_NotKnown)
            print('Number of', ethnicity, ' :', Sum, 'corresponding to %0.2f' % Prop, '%')

    print('\n')
    # --------------------------------------------
    # Deprivation Quintile: already included in updated datasets
#     ColumnList = ['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)']
#     df_IMD_Sub = df_IMD[ColumnList]

#     df_Table1 = pd.merge(df_Table1, df_IMD_Sub, left_on='LSOA_2011_Code', right_on='LSOA code (2011)', how='left')
#     df_Table1 = df_Table1.drop(columns=['LSOA code (2011)'])
#     df_Table1 = df_Table1.rename(columns={'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)':'IMD Deciles'})
#     df_Table1.head()

    if 'IMD_Decile' in df_Table1.columns:
        Decile_List = range(1,10+1)
        for k in range(5):
            Decile_Max = Decile_List[k]*2
            Decile_Min = Decile_Max-1

            Filter_Quintiles = (df_Table1["IMD_Decile"]>=Decile_Min) & (df_Table1["IMD_Decile"]<=Decile_Max)
            N_Patients_Quintile = df_Table1[Filter_Quintiles].shape[0]
            N_Patients_Quintile_Missing = df_Table1.loc[Filter_Quintiles, 'IMD_Decile'].isnull().sum()
            Prop = 100*N_Patients_Quintile/(N_Patients-N_Patients_Quintile_Missing)
            print('IMD Deciles:', Decile_Min, '-', Decile_Max)
            print('\t Number of patients:', N_Patients_Quintile, 'corresponding to %0.2f' % Prop, '%')

        N_MissingIMD = df_Table1["IMD_Decile"].isnull().sum()
        Prop_Missing = 100*N_MissingIMD/(N_Patients-N_MissingIMD)
        print('Number of missing IMD:', N_MissingIMD, 'corresponding to %0.2f' % Prop_Missing, '%')

    print('\n')
    # --------------------------------------------
    # Number of readmissions
    df_Count = df_ini.copy()
#     df_UniqueSpells = df_Count.sort_values(by=['HESID', 'EPIend']).drop_duplicates(subset=['P_Spell_ID'], keep='first')[['HESID','P_Spell_ID', 'SpellLevel_TimeDiff']]
    df_UniqueSpells = df_Count.sort_values(by=['HESID', 'EPIend']).drop_duplicates(subset=['P_Spell_ID'], keep='first')
    df_UniqueSpells.loc[:, 'SpellLevel_TimeDiff'] = (df_UniqueSpells.loc[:, 'Admission_date'] - df_UniqueSpells.loc[:, 'FirstEpisode_Discharge_date']).dt.days

    Filter_3Years = df_UniqueSpells.loc[:, 'SpellLevel_TimeDiff']<=3*365
    df_NumberSpells = df_UniqueSpells.loc[Filter_3Years, :].groupby(by='HESID').agg('count').reset_index().rename(columns={'P_Spell_ID':'Number Spells'})
    df_NumberSpells.loc[:, 'Number Readmissions'] = df_NumberSpells.loc[:, 'Number Spells']-1
    Readmissions_Describe = df_NumberSpells['Number Readmissions'].describe()
    print('Number of readmissions within 3 years:', Readmissions_Describe['50%'],
          '(', Readmissions_Describe['25%'], '-', Readmissions_Describe['75%'], ')')

    print('\n')
    # --------------------------------------------
    # most common specialty
    Describe = df_Table1["Main_Specialty_Description"].describe()
    print(Describe['top'], ': Proportion %0.2f' % (100*Describe['freq']/Describe['count']), '%' )
    Filter_SecondMostCommon = (df_Table1["Main_Specialty_Description"] != Describe['top'])
    Describe_SecondMostCommon = df_Table1.loc[Filter_SecondMostCommon, 'Main_Specialty_Description'].describe()
    print(Describe_SecondMostCommon['top'],
          ': Proportion %0.2f' % (100*Describe_SecondMostCommon['freq']/Describe_SecondMostCommon['count']), '%' )
    
    print('\n')
    # --------------------------------------------
    # median number of episode per sell
    NumberEpisodes = df_dementia_3.groupby(by=['P_Spell_ID'])['EPIstart'].agg('count').describe()
    print('Median number of episodes per spell:', NumberEpisodes['50%'],
          '(', NumberEpisodes['25%'], '-', NumberEpisodes['75%'], ')')
    
    print('\n')
    # --------------------------------------------
    # median number of spells per patient
    NumberSpells = df_UniqueSpells.groupby(by=['HESID'])['P_Spell_ID'].agg('count').describe()
    print('Median number of spells per patient:', NumberSpells['50%'],
          '(', NumberSpells['25%'], '-', NumberSpells['75%'], ')')
    
print('--------------------------------------------------')
print('+ \033[1m Dementia \033[0m')
ProfileFirstSpells(df_dementia_3)
print('\n')

print('--------------------------------------------------')
print('+ \033[1m Autism \033[0m')
ProfileFirstSpells(df_autism_3)
print('\n')

print('--------------------------------------------------')
print('+ \033[1m Diabetes \033[0m')
ProfileFirstSpells(df_diabetes_3)
```

## Find ICD10 codes across all episodes within spells

```python


# Converting List of ICD-10 codes to string without duplicates
# --------------------------------------------------
def ICD10Chain_ToList(List):
    List2 = list( dict.fromkeys(List) )
    List2 = [x for x in List2 if x!='-1']
    String = '~'.join(List2)

    return String
# --------------------------------------------------


# Add SpellLevel_ICD10 feature to dataframe and keep last episode of each spell
# --------------------------------------------------
def SpellLevel_ICD10(df_ini):
    df = df_ini.copy()

    # Find ICD-10 codes across spells
    ICD10_Spells = df.groupby(by=['P_Spell_ID'])['diagnosis_group'].agg('~'.join).reset_index()
    ICD10_Spells.loc[:, 'diagnosis_group'] = ICD10_Spells.loc[:, 'diagnosis_group'].apply(lambda x: ICD10Chain_ToList( x.split('~') ) )

    df_Merged = pd.merge(df, ICD10_Spells, on='P_Spell_ID', how='left').rename(columns={'diagnosis_group_x':'diagnosis_group',
                                                                                        'diagnosis_group_y':'SpellLevel_ICD10'})
    # Keep only last episode of each spell
    df_Merged = df_Merged.drop_duplicates(subset=['P_Spell_ID'], keep='last')
    
    return df_Merged
# --------------------------------------------------

df_dementia_SpellLevel = SpellLevel_ICD10(df_dementia_3)
df_autism_SpellLevel = SpellLevel_ICD10(df_autism_3)
df_diabetes_SpellLevel = SpellLevel_ICD10(df_diabetes_3)

displayDataFrame(df_dementia_SpellLevel.head())
displayDataFrame(df_autism_SpellLevel.head())
displayDataFrame(df_diabetes_SpellLevel.head())
```

```python

# Find mentions of mandatory codes in CodeList in ICDfeature_Name
# --------------------------------------------------
def SpellLevel_FindMandatoryCode(df_ini, ICDfeature_Name, ConditionName, CodeList):
    df = df_ini
    
    Codes = '|'.join(CodeList)

    # Look for ANY mention of mandatory code
    Filter = df[ICDfeature_Name].str.contains(Codes)
    df.loc[:, "SpellLevel_ICD10_" + ConditionName] = np.where(Filter, 1, 0)

    return df
# --------------------------------------------------

# Mandatory codes for dementia
CodesDementia_Categories = "F00|F01|F02"
CodesDementia_Unspecified = "F03|F051"
Codes_Dementia_SubCategories = "F000|F001|F002|F009|F010|F011|F012|F013|F018|F019|F020|F021|F022|F023|F024|F028"
Codes_Dementia_Supp = "G300|G301|G308|G309" # No G302, G300 instead!
CodesDementia = CodesDementia_Categories + "|" + CodesDementia_Unspecified + "|" + Codes_Dementia_SubCategories + "|" + Codes_Dementia_Supp
CodesDementia_List = CodesDementia.split("|")
Parkinson_List = ["G20"] 
ParkinsonsDementia_List = ["F023"]

# Mandatory codes for autism
Codes_LearningDisabilities = ['F7' + i for i in ['0', '1', '2', '3', '8', '9']]
CodesAutism_List = ["F840", "F841", "F845"]

# Mandatory codes for diabetes
CodesDiabetes_List = ["E10", "E11", "E14"]
CodesDiabetesTypeII_List = ["E11"]

# Look for mandatory codes
df_dementia_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_dementia_SpellLevel, "SpellLevel_ICD10", "Dementia", CodesDementia_List)
df_dementia_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_dementia_SpellLevel_2, "SpellLevel_ICD10", "Parkinson", Parkinson_List)
df_dementia_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_dementia_SpellLevel_2, "SpellLevel_ICD10", "ParkinsonsDementia", ParkinsonsDementia_List)

df_autism_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_autism_SpellLevel, "SpellLevel_ICD10", "Autism", CodesAutism_List)
df_autism_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_autism_SpellLevel_2, "SpellLevel_ICD10", "LearningDisabilities", Codes_LearningDisabilities)

df_diabetes_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_diabetes_SpellLevel, "SpellLevel_ICD10", "Diabetes", CodesDiabetes_List)
df_diabetes_SpellLevel_2 = SpellLevel_FindMandatoryCode(df_diabetes_SpellLevel_2, "SpellLevel_ICD10", "DiabetesTypeII", CodesDiabetesTypeII_List)


SummaryDataFrame(df_dementia_SpellLevel_2, 'Dementia')
SummaryDataFrame(df_autism_SpellLevel_2, 'Autism')
SummaryDataFrame(df_diabetes_SpellLevel_2, 'Diabetes')

```

```python

# --------------------------------------------------
def SpellLevel_FindErrors(df_ini, ConditionName):
    df = df_ini
    
    # Exclude all spells with discharge date > end cut-off-date
    Filter_NotEnd = (df.loc[:, 'Discharge_date']<=pd.to_datetime('2021-03-31', format='%Y-%m-%d'))
    df = df.loc[Filter_NotEnd, :]

    # Exclude all spells with discharge date < start cut-off-date
    Filter_AfterFirstTwoYears = (df.loc[:, 'Discharge_date']>=pd.to_datetime('2013-04-01', format='%Y-%m-%d'))
    df = df.loc[Filter_AfterFirstTwoYears, :]
    
    Filter_Code = df["SpellLevel_ICD10_"+ConditionName] == 1
    df_Code = df.loc[Filter_Code, :]
    df_NoCode = df.loc[~Filter_Code, :]

    # Keep first occurence of mandatory code
    df_CodeFirst = df_Code.sort_values(by=['HESID', 'Admission_date', 'Discharge_date']).drop_duplicates(subset=['HESID'], keep='first')

    ColumnList = ['HESID', 'P_Spell_ID',
                  'Admission_date', 'Discharge_date',
                  'ProvCode', 'POD', 'Main_Specialty_Description']
    
    df = pd.merge( df, df_CodeFirst[ColumnList], how='left', on='HESID' )
    
    ColumnRename_Original = [x for x in df.columns if '_x' in x]
    ColumnRename_First = [x for x in df.columns if '_y' in x]
    ColumnRename = ColumnRename_Original + ColumnRename_First
    
    d_rename = dict(zip(ColumnRename, ColumnRename))
    for feature in ColumnRename:
        if '_x' in feature:
            d_rename[feature] = feature[:-2]
        elif 'y' in feature:
            d_rename[feature] = 'FirstSpell_' + feature[:-2]
    
    df = df.rename(columns=d_rename)
    
    # Flag first spell
    df_CodeFirst.loc[:, 'SpellLevel_FirstDiagnosis'] = 1
    ColumnList = ['P_Spell_ID', 'SpellLevel_FirstDiagnosis']
    df = pd.merge(df, df_CodeFirst[ColumnList], on='P_Spell_ID', how='left')
    df.loc[:, 'SpellLevel_FirstDiagnosis'] = df.loc[:, 'SpellLevel_FirstDiagnosis'].fillna(0)
    df = df.astype({'SpellLevel_FirstDiagnosis':int})
        
    # Flag subsequent episodes
    Filter_SubsequentEpisode = (df['FirstSpell_Discharge_date']<=df['Discharge_date']) & (df.loc[:, 'SpellLevel_FirstDiagnosis']==0)
    df.loc[:, 'SubsequentSpell'] = np.where(Filter_SubsequentEpisode, 1, 0)
    
    # Find errors in dataset
    Filter_Error = (df["SpellLevel_ICD10_"+ConditionName]==0) & (df['FirstSpell_Discharge_date']<df['Discharge_date'])
    df.loc[:, 'SpellLevel_Error_'+ConditionName] = np.where(Filter_Error, 1, 0)

    # Remove features at episode level
    Features_SpellLevel = [ 'P_Spell_ID', 'HESID', 'LSOA_2011_Code', 'Admission_date',
                           'Discharge_date', 'EPIstart', 'EPIend', 'diagnosis_group', 'Spell_Los',
                           'sex', 'age_of_patient', 'Ethnicity', 'IMD_Score', 'IMD_Decile',
                           'HFRS_Score', 'HFRS_Band', 'ProvCode', 'Main_Specialty_Description',
                           'POD', 'MORT', 'Read30', 'Read90']
    
    Features_SpellLevel = Features_SpellLevel + [feature for feature in df.columns if 'SpellLevel_' in feature]
    Features_SpellLevel = Features_SpellLevel + [feature for feature in df.columns if 'FirstSpell_' in feature]
    Features_SpellLevel = Features_SpellLevel + ['SubsequentSpell']
    
    df = df.loc[:, Features_SpellLevel]
    
    df = df.sort_values(by=['HESID', 'Admission_date', 'Discharge_date'])
    
    return df
# --------------------------------------------------

df_dementia_SpellLevel_3 = SpellLevel_FindErrors(df_dementia_SpellLevel_2, 'Dementia')
df_autism_SpellLevel_3 = SpellLevel_FindErrors(df_autism_SpellLevel_2, 'Autism')
df_diabetes_SpellLevel_3 = SpellLevel_FindErrors(df_diabetes_SpellLevel_2, 'Diabetes')

SummaryDataFrame(df_dementia_SpellLevel_3, 'Dementia')
SummaryDataFrame(df_autism_SpellLevel_3, 'Autism')
SummaryDataFrame(df_diabetes_SpellLevel_3, 'Diabetes')
```

```python

def FindErrorsWithSubCodes(df_ini, ConditionName):
    df = df_ini.copy()
    
    # Keep subsequent spells
    Filter_SubsequentSpells = df.loc[:, 'SubsequentSpell']==1
    df = df.loc[Filter_SubsequentSpells, :]
    
    if ConditionName == 'Dementia':
    # ---------- Dementia
        print('+ \033[1m DEMENTIA \033[0m ')
        # Dementia: errors at spell level
        TotalNumber_Spells = df.shape[0]
        Errors_Number = df.loc[:, 'SpellLevel_Error_Dementia'].sum()
        Error_Prop = 100 * Errors_Number / TotalNumber_Spells
        print('Number of errors at spell level:', Errors_Number, 'corresponding to %0.2f' % Error_Prop, '%')

        # Dementia: Parkinson's Disease (G20)
        G20_Number = df.loc[:, 'SpellLevel_ICD10_Parkinson'].sum()
        G20_Prop = 100 * G20_Number / TotalNumber_Spells
        print('Number of spells with G20 code:', G20_Number, 'corresponding to %0.2f' % G20_Prop, '%')

        # G20 and errors
        Filter_G20 = (df.loc[:, 'SpellLevel_ICD10_Parkinson']==1) 
        NumberErrors_WithG20 = df.loc[Filter_G20, 'SpellLevel_Error_Dementia'].sum()
        PropErrors_WithG20 = 100 * NumberErrors_WithG20 / G20_Number
        print('Number of errors in spells WITH G20 code:', NumberErrors_WithG20, 'corresponding to %0.2f' % PropErrors_WithG20, '%')
        NumberErrors_WithoutG20 = df.loc[~Filter_G20, 'SpellLevel_Error_Dementia'].sum()
        PropErrors_WithoutG20 = 100 * NumberErrors_WithoutG20 / (df.loc[~Filter_G20, :].shape[0])
        print('Number of errors in spells WITHOUT G20 code:', NumberErrors_WithoutG20,
              'corresponding to %0.2f' % PropErrors_WithoutG20, '%')


    if ConditionName == 'Diabetes':
    # ---------- Diabetes
        print('+ \033[1m DIABETES \033[0m ')
        # Autism: errors at spell level
        TotalNumber_Spells = df.shape[0]
        Errors_Number = df.loc[:, 'SpellLevel_Error_Diabetes'].sum()
        Error_Prop = 100 * Errors_Number / TotalNumber_Spells
        print('Number of errors at spell level:', Errors_Number, 'corresponding to %0.2f' % Error_Prop, '%')

        # Type 2 Diabetes: (E11)
        E11_Number = df.loc[:, 'SpellLevel_ICD10_DiabetesTypeII'].sum()
        E11_Prop = 100 * E11_Number / TotalNumber_Spells
        print('Number of spells with E11 code:', E11_Number, 'corresponding to %0.2f' % E11_Prop, '%')

        # E11 and errors
        Filter_E11 = (df.loc[:, 'SpellLevel_ICD10_DiabetesTypeII']==1) 
        NumberErrors_WithE11 = df.loc[Filter_E11, 'SpellLevel_Error_Diabetes'].sum()
        PropErrors_WithE11 = 100 * NumberErrors_WithE11 / E11_Number
        print('Number of errors in spells WITH E11 code:', NumberErrors_WithE11,
              'corresponding to %0.2f' % PropErrors_WithE11, '%')
        NumberErrors_WithoutE11 = df.loc[~Filter_E11, 'SpellLevel_Error_Diabetes'].sum()
        PropErrors_WithoutE11 = 100 * NumberErrors_WithoutE11 / (df.loc[~Filter_E11, :].shape[0])
        print('Number of errors in spells WITHOUT E11 code:', NumberErrors_WithoutE11,
              'corresponding to %0.2f' % PropErrors_WithoutE11, '%')

    if ConditionName == 'Autism':
    # ---------- Autism
        print('+ \033[1m AUTISM \033[0m ')
        N_Errors = df.loc[:, 'SpellLevel_Error_Autism'].sum()
        N_Total = df.shape[0]
        Prop_Errors = 100 * N_Errors / N_Total
        print('Number of errors at spell level:', N_Errors, 'corresponding to %0.2f' % Prop_Errors, '%')
#         Codes_LearningDisabilities = ['F7' + i + x for i in ['0', '1', '2', '3', '8', '9'] for x in ['0', '1', '8', '9']]
#         Codes_LearningDisabilities = Codes_LearningDisabilities + ['F8' + i + x for i in ['0', '1'] for x in ['0', '1', '2', '3', '8', '9']]
#         Codes_LearningDisabilities = Codes_LearningDisabilities + ['F82', 'F83'] + ['F840', 'F841', 'F845'] + ['F842', 'F844', 'F848', 'F849', 'F88', 'F89']
        Codes_LearningDisabilities = ['F7' + i for i in ['0', '1', '2', '3', '8', '9']]
        Codes_LearningDisabilities_str = '|'.join(Codes_LearningDisabilities)
        Filter_Codes_LearningDisabilities = df.loc[:, 'SpellLevel_ICD10'].str.contains(Codes_LearningDisabilities_str)
        N_LearningDisabilities = df.loc[Filter_Codes_LearningDisabilities, :].shape[0]
        Prop_LearningDisabilities = 100 * N_LearningDisabilities / N_Total
        print('Number of spells with learning disability codes:', N_LearningDisabilities,
              'corresponding to %0.2f' % Prop_LearningDisabilities, '%')
        N_Errors_WithLearningDisabilityCodes = df.loc[Filter_Codes_LearningDisabilities, 'SpellLevel_Error_Autism'].sum()
        Prop_Errors_WithLearningDisabilityCodes = 100 * N_Errors_WithLearningDisabilityCodes / (df.loc[Filter_Codes_LearningDisabilities, :].shape[0])
        print('Number of errors in spells WITH learning disability codes:', N_Errors_WithLearningDisabilityCodes,
              'corresponding to %0.2f' % Prop_Errors_WithLearningDisabilityCodes, '%')
        N_Errors_WithoutLearningDisabilityCodes = df.loc[~Filter_Codes_LearningDisabilities, 'SpellLevel_Error_Autism'].sum()
        Prop_Errors_WithoutLearningDisabilityCodes = 100 * N_Errors_WithoutLearningDisabilityCodes / (df.loc[~Filter_Codes_LearningDisabilities, :].shape[0])
        print('Number of errors in spells WITHOUT learning disability codes:', N_Errors_WithoutLearningDisabilityCodes,
              'corresponding to %0.2f' % Prop_Errors_WithoutLearningDisabilityCodes, '%')

FindErrorsWithSubCodes(df_dementia_SpellLevel_3, 'Dementia')
FindErrorsWithSubCodes(df_autism_SpellLevel_3, 'Autism')
FindErrorsWithSubCodes(df_diabetes_SpellLevel_3, 'Diabetes')
```
