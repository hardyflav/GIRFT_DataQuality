### Engineer features of interest

```python

# Feature Engineering
# --------------------------------------------------
def FeatureEngineering(df_ini, df_TrustRegions, ConditionName):
    df = df_ini.copy()
    
    DateFeatures = ['Admission_date', 'Discharge_date', 'EPIstart', 'EPIend',
                    'FirstSpell_Admission_date', 'FirstSpell_Discharge_date']
    for datefeature in DateFeatures:
        df.loc[:, datefeature] = pd.to_datetime(df.loc[:, datefeature], format='%Y-%m-%d')

    # Time between current EPIend and first instance of dementia code
    df.loc[:, 'TimeDiff_FirstSpell'] = (df['Admission_date'] - df['FirstSpell_Discharge_date']).dt.days

    # Whether admission in different trust, compared to original one
    Filter_ChangeProv = df['FirstSpell_ProvCode'] != df['ProvCode']
    df.loc[:, 'Change_ProvCode'] = np.where(Filter_ChangeProv, 1, 0)

    # Whether admission in different region, compared to original one
    df = pd.merge(df, df_TrustRegions, on='ProvCode', how='left')
    df = pd.merge(df, df_TrustRegions, left_on='FirstSpell_ProvCode', right_on='ProvCode', how='left')
    df = df.rename(columns={'region_x':'Region','region_y':'FirstSpell_Region',
                            'ProvCode_x':'ProvCode'})
    df = df.drop(columns=['ProvCode_y'])
    df.loc[:,'Region'] = df.loc[:,'Region'].fillna('Non-NHS provider')
    df.loc[:,'FirstSpell_Region'] = df.loc[:,'FirstSpell_Region'].fillna('Non-NHS provider')

    Filter_ChangeRegion = df['FirstSpell_Region'] != df['Region']
    df.loc[:, 'Change_Region'] = np.where(Filter_ChangeRegion, 1, 0)
    
    Filter_SpecialtyChange = (df['Main_Specialty_Description'] != df['FirstSpell_Main_Specialty_Description'])
    df.loc[:, 'Change_Main_Specialty_Description'] = np.where(Filter_SpecialtyChange, 1, 0)

    # Missing age (999)
    Filter_MissingAge = df['age_of_patient'] == 999
    df.loc[:, 'Missing_Age'] = np.where(Filter_MissingAge, 1, 0)
    
    # Correct missing age
    MedianAge = df.loc[~Filter_MissingAge, 'age_of_patient'].median()
    df.loc[Filter_MissingAge, 'age_of_patient'] = MedianAge

    # Missing IMD scores
    Filter_MissingIMD = df['IMD_Score'].isnull()
    df.loc[:, 'Missing_IMD_Score'] = np.where(Filter_MissingIMD, 1, 0)
    df['IMD_Score'] = df['IMD_Score'].fillna(df['Missing_IMD_Score'].mean())
    
    # Correct HFRS_Score=='None' in autism
    Filter_None = df.loc[:, 'HFRS_Score']=='None'
    df.loc[Filter_None, 'HFRS_Score'] = 0
    df.loc[:, 'HFRS_Score'] = pd.to_numeric(df.loc[:, 'HFRS_Score'], downcast='float')
    
    # Correct upper/lower case  values
    df.loc[:, 'Ethnicity'] = df.loc[:, 'Ethnicity'].str.lower()
    df.loc[:, 'Main_Specialty_Description'] = df.loc[:, 'Main_Specialty_Description'].str.lower()
    df.loc[:, 'FirstSpell_Main_Specialty_Description'] = df.loc[:, 'FirstSpell_Main_Specialty_Description'].str.lower()
        
    # Elective / Emergency surgeries
    Filter_POD_Elective = df.loc[:, 'POD'].str.contains('EL|DC|RA')
    df.loc[:, 'POD_Elective'] = np.where(Filter_POD_Elective, 1, 0)
    Filter_POD_Emergency = df.loc[:, 'POD'].str.contains('EM|NE')
    df.loc[:, 'POD_Emergency'] = np.where(Filter_POD_Emergency, 1, 0)
    Filter_POD_Missing = df.loc[:, 'POD'].isnull()
    df.loc[:, 'Missing_POD'] = np.where(Filter_POD_Missing, 1, 0)
    
    # NHS / Private Providers
#     PrivateProv_List = list(df_TrustRegions[df_TrustRegions['region']=='Non-NHS provider']['ProvCode'].values)
#     PrivateProv_String = '|'.join(PrivateProv_List)
#     Filter_NonNHSProv = df.loc[:, 'ProvCode'].str.contains(PrivateProv_String)
#     df.loc[:, 'ProvCode_PrivateProv'] = np.where(Filter_NonNHSProv, 1, 0)
#     Filter_NonNHSProvFirst = df.loc[:, 'FirstSpell_ProvCode'].str.contains(PrivateProv_String)
#     df.loc[:, 'FirstSpell_ProvCode_PrivateProv'] = np.where(Filter_NonNHSProvFirst, 1, 0)

    Filter_NHS = df.loc[:, 'ProvCode'].str.startswith('R')
    df.loc[:, 'ProvCode_PrivateProv'] = np.where(Filter_NHS, 0, 1)
    Filter_NHSFirst = df.loc[:, 'FirstSpell_ProvCode'].str.startswith('R')
    df.loc[:, 'FirstSpell_ProvCode_PrivateProv'] = np.where(Filter_NHSFirst, 0, 1)
    
    # Change between NHS-Private providers
    Filter_ChangeTypeProv = df.loc[:, 'FirstSpell_ProvCode_PrivateProv'] != df.loc[:, 'ProvCode_PrivateProv']
    df.loc[:, 'Change_ProvCode_Type'] = np.where(Filter_ChangeTypeProv, 1, 0)
    
    # Zero-day LOS
    Filter_0LOS = df.loc[:, 'Spell_Los']==0
    df.loc[:, 'Spell_Los_0'] = np.where(Filter_0LOS, 1, 0)
        
    # Reorganising columns
    ColumnList_SpellLevel = [ feature for feature in df.columns if 'SpellLevel_' in feature ] + ['SubsequentSpell']
    ColumnList_FirstSpell = [ feature for feature in df.columns if 'FirstSpell_' in feature ]
    ColumnList_Change = [ feature for feature in df.columns if 'Change_' in feature ]
    ColumnList_Other = [ feature for feature in df.columns if not(feature in ColumnList_SpellLevel+ColumnList_FirstSpell+ColumnList_Change) ]
    ColumnList = ColumnList_Other + ColumnList_FirstSpell + ColumnList_Change + ColumnList_SpellLevel
    ColumnList = [feature for feature in ColumnList if feature!='SpellLevel_Error_'+ConditionName] + ['SpellLevel_Error_'+ConditionName]
    
    df = df.loc[:, ColumnList]
    
    if ('Unnamed: 0' in df.columns):
        df = df.drop(columns=['Unnamed: 0'])
    
    return df

# --------------------------------------------------

df_dementia = FeatureEngineering(df_dementia_ini, df_TrustRegions, 'Dementia')
SummaryDataFrame(df_dementia, 'Dementia')

df_autism = FeatureEngineering(df_autism_ini, df_TrustRegions, 'Autism')
SummaryDataFrame(df_autism, 'Autism')

df_diabetes = FeatureEngineering(df_diabetes_ini, df_TrustRegions, 'Diabetes')
SummaryDataFrame(df_diabetes, 'Diabetes')
```

## Identify Coding Errors at Spell Level

### Add ICD-10 codes across spells

```python

# Converting List of ICD-10 codes to string without duplicates
# --------------------------------------------------
def ICD10Chain_ToList(List):
    List2 = list( dict.fromkeys(List) )
    List2 = [x for x in List2 if x!='-1']
    String = '~'.join(List2)

    return String
# --------------------------------------------------


df_dementia_ini.head()
df = df_dementia_ini.loc[0:10,:].copy()

# Add SpellLevel_ICD10 feature to dataframe
# --------------------------------------------------
def SpellLevel_ICD10(df_ini):
    
    df = df_ini

    # Find ICD-10 codes across spells
    ICD10_Spells = df.groupby(by=['P_Spell_ID'])['diagnosis_group'].agg('~'.join).reset_index()
    ICD10_Spells.loc[:, 'diagnosis_group'] = ICD10_Spells.loc[:, 'diagnosis_group'].apply(lambda x: ICD10Chain_ToList( x.split('~') ) )

    df_Merged = pd.merge(df, ICD10_Spells, on='P_Spell_ID', how='left').rename(columns={'diagnosis_group_x':'diagnosis_group',
                                                                                        'diagnosis_group_y':'SpellLevel_ICD10'})
    
    return df_Merged
# --------------------------------------------------

df_dementia_SpellLevel = SpellLevel_ICD10(df_dementia)
df_autism_SpellLevel = SpellLevel_ICD10(df_autism)
df_diabetes_SpellLevel = SpellLevel_ICD10(df_diabetes)

df_autism_SpellLevel.head()
```

### Find mandatory codes

```python

# Find mentions of mandatory codes in CodeList in ICDfeature_Name
# --------------------------------------------------
def SpellLevel_FindMandatoryCode(df_ini, ICDfeature_Name, ConditionName, CodeList):
    df = df_ini
    
    Codes = '|'.join(CodeList)

    # Look for ANY mention of mandatory code
    Filter = df[ICDfeature_Name].str.contains(Codes)
    df.loc[:, "SpellLevel_ICD10_" + ConditionName] = np.where(Filter, 1, 0)

    return df
# --------------------------------------------------

# Mandatory codes for dementia
CodesDementia_Categories = "F00|F01|F02"
CodesDementia_Unspecified = "F03|F051"
Codes_Dementia_SubCategories = "F000|F001|F002|F009|F010|F011|F012|F013|F018|F019|F020|F021|F022|F023|F024|F028"
Codes_Dementia_Supp = "G301|G302|G308|G309"
CodesDementia = CodesDementia_Categories + "|" + CodesDementia_Unspecified + "|" + Codes_Dementia_SubCategories + "|" + Codes_Dementia_Supp
CodesDementia_List = CodesDementia.split("|")

# Mandatory codes for autism
CodesAutism_List = ['F84']

# Mandatory codes for diabetes
CodesDiabetes_List = ['E11']

# Look for mandatory codes
df_dementia_SpellLevel2 = SpellLevel_FindMandatoryCode(df_dementia_SpellLevel, 'SpellLevel_ICD10', 'Dementia', CodesDementia_List)
df_autism_SpellLevel2 = SpellLevel_FindMandatoryCode(df_autism_SpellLevel, "SpellLevel_ICD10", "Autism", CodesAutism_List)
df_diabetes_SpellLevel2 = SpellLevel_FindMandatoryCode(df_diabetes_SpellLevel, "SpellLevel_ICD10", "Diabetes", CodesDiabetes_List)

df_dementia_SpellLevel2.head()
```

```python

# Keep last episode for unique spells
df_dementia_SpellLevel2 = df_dementia_SpellLevel2.drop_duplicates(subset=['P_Spell_ID'], keep='last')
df_autism_SpellLevel2 = df_autism_SpellLevel2.drop_duplicates(subset=['P_Spell_ID'], keep='last')
df_diabetes_SpellLevel2 = df_diabetes_SpellLevel2.drop_duplicates(subset=['P_Spell_ID'], keep='last')

# Find missing mandatory codes at spell level
# --------------------------------------------------
def SpellLevel_FindErrors(df_ini, ConditionName):
    df = df_ini
    
    Filter_Code = df["SpellLevel_ICD10_"+ConditionName] == 1
    df_Code = df[Filter_Code]
    df_NoCode = df[~Filter_Code]
    
    # Keep first occurence of mandatory code
    df_CodeFirst = df_Code.sort_values(by=['HESID', 'Admission_date']).drop_duplicates(subset=['HESID'], keep='first')

    ColumnList = ['HESID', 'Admission_date', 'ProvCode']
    df = pd.merge( df, df_CodeFirst[ColumnList], how='left', on='HESID' )
    df = df.rename(columns={'Admission_date_x':'Admission_date', 'Admission_date_y':'SpellLevel_Admission_date_First',
                            'ProvCode_x':'ProvCode',
                            'ProvCode_y':'SpellLevel_ProvCode_First'})
                
    # Find errors in dataset
    Filter_Error = (df["SpellLevel_ICD10_"+ConditionName]==0) & (df['SpellLevel_Admission_date_First']<df['Admission_date'])
    df.loc[:, 'SpellLevel_Error_'+ConditionName] = np.where(Filter_Error, 1, 0)

    df = df.sort_values(by=['HESID', 'Admission_date'])
    
    return df
# --------------------------------------------------

df_dementia_SpellLevel3 = SpellLevel_FindErrors(df_dementia_SpellLevel2, 'Dementia')
df_autism_SpellLevel3 = SpellLevel_FindErrors(df_autism_SpellLevel2, 'Autism')
df_diabetes_SpellLevel3 = SpellLevel_FindErrors(df_diabetes_SpellLevel2, 'Diabetes')

df_dementia_SpellLevel3.head()
```

```python

# Count missing mandatory codes at spell level
# --------------------------------------------------
def SpellLevel_CountErrors(df_ini, ConditionName):
    df = df_ini
    
    # Find subsequent episodes
    Filter_SubsequentSpells = (df.loc[:, 'Admission_date']>df.loc[:, 'SpellLevel_Admission_date_First'])
    df_SubsequentSpells = df.loc[Filter_SubsequentSpells, :]

    Sum = df_SubsequentSpells.loc[:, 'SpellLevel_Error_'+ConditionName].sum()
    Count = df_SubsequentSpells.loc[:, 'SpellLevel_Error_'+ConditionName].shape[0]
    Prop = 100 * Sum / Count
    
    return Sum, Count, Prop
# --------------------------------------------------

Sum_dementia, Count_dementia, Prop_dementia = SpellLevel_CountErrors(df_dementia_SpellLevel3, 'Dementia')
print('-----------------------------')
print('DEMENTIA')
print('Number of errors:', Sum_dementia)
print('Number of subsequent SPELLS where mandatory codes should be used:', Count_dementia)
print('Proportion of errors: %0.2f' % Prop_dementia, '%')

Sum_autism, Count_autism, Prop_autism = SpellLevel_CountErrors(df_autism_SpellLevel3, 'Autism')
print('-----------------------------')
print('AUTISM')
print('Number of errors:', Sum_autism)
print('Number of subsequent SPELLS where mandatory codes should be used:', Count_autism)
print('Proportion of errors: %0.2f' % Prop_autism, '%')

Sum_diabetes, Count_diabetes, Prop_diabetes = SpellLevel_CountErrors(df_diabetes_SpellLevel3, 'Diabetes')
print('-----------------------------')
print('DIABETES')
print('Number of errors:', Sum_diabetes)
print('Number of subsequent SPELLS where mandatory codes should be used:', Count_diabetes)
print('Proportion of errors: %0.2f' % Prop_diabetes, '%')


```

```python

# Link ethinicity data for Dementia dataset
# --------------------------------------------------
def Dementia_LinkEthnicityIMD(df_ini, df_Ethnicity_ini):
    df = df_ini.copy()
    df_Ethnicity = df_Ethnicity_ini.copy()
    ConditionName = 'Dementia'

    # Dropping unwanted index columns | Renaming columns
    ColumnsUnnamed = ['Unnamed: 0', 'Unnamed: 0.1']
    for col in ColumnsUnnamed:
        if col in df.columns: 
            df = df.drop(columns=[col])

    df_Ethnicity = df_Ethnicity.rename(columns={'HESID1':'HESID'})

    # Convert date features
    DateFeatures = ['Admission_date', 'Discharge_date', 'EPIstart', 'EPIend', 'EPIend_First']
    for datefeature in DateFeatures:
        df.loc[:, datefeature] = pd.to_datetime(df.loc[:, datefeature], format='%Y-%m-%d')
        
    # Find discharge date of first spell
    Filter_Diagnosis = df.loc[:, 'SpellLevel_ICD10_'+ConditionName]==1
    df_First = df.loc[Filter_Diagnosis, :].sort_values(by=['Admission_date', 'EPIstart', 'EPIend']).drop_duplicates(subset=['HESID'], keep='first')
    df = pd.merge(df, df_First[['HESID', 'Discharge_date']], on='HESID', how='left')
    df = df.rename(columns={'Discharge_date_x':'Discharge_date',
                            'Discharge_date_y':'SpellLevel_Discharge_date_First'})    

    # Ethnicity column
    ColumnList = ['HESID', 'Ethnicity', 'HFRS_Score', 'HFRS_Band']
    df_Ethnicity = df_Ethnicity.loc[:, ColumnList].drop_duplicates(subset=['HESID'], keep='last')
    df_Merged = pd.merge(df, df_Ethnicity, on='HESID', how='left')

    # Correct IMD scores: AFTER 2013
    ColumnList = ['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Score',
                  'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)']
    df_IMD_Sub = df_IMD.loc[:, ColumnList].rename(columns={'LSOA code (2011)':'LSOA_2011_Code',
                                                           'Index of Multiple Deprivation (IMD) Score':'IMD Score',
                                                           'Index of Multiple Deprivation (IMD) Decile (where 1 is most deprived 10% of LSOAs)':'IMD Decile'})
    df_IMD_Sub = df_IMD_Sub.drop_duplicates(subset=['LSOA_2011_Code'], keep='first')

    df_Merged2 = pd.merge(df_Merged, df_IMD_Sub, on='LSOA_2011_Code', how='left')
    df_Merged2 = df_Merged2.drop(columns=['IMD Score_x']).rename(columns={'IMD Score_y':'IMD Score'})

    # Exclude all spells with discharge date < cut-off-date
    Filter_AfterFirstTwoYears = (df_Merged2.loc[:, 'Discharge_date']>pd.to_datetime('2013-04-01', format='%Y-%m-%d'))
    df_Merged3 = df_Merged2.loc[Filter_AfterFirstTwoYears, :]
    
    # Find new date for first spell with code: NewFirst
    Filter_Code = (df_Merged3.loc[:, 'SpellLevel_ICD10_'+ConditionName]==1)
    df_NewFirst = df_Merged3.loc[Filter_Code, :].sort_values(by=['Admission_date',
                                                                 'EPIstart', 'EPIend']).drop_duplicates(subset=['HESID'], keep='first')
    ColumnList = ['HESID', 'Discharge_date', 'Admission_date']
    df_Merged3 = pd.merge(df_Merged3, df_NewFirst.loc[:, ColumnList], on='HESID', how='left').rename(columns={'Discharge_date_x':'Discharge_date',
                                                                                                              'Discharge_date_y':'NewFirstSpell_Discharge_date',
                                                                                                              'Admission_date_x':'Admission_date',
                                                                                                              'Admission_date_y':'NewFirstSpell_Admission_date'})
    # Exclude (First spell < cut-off-date) & (spell < NewFirst)
    Filter_NoCorrectSpellAfterTwoYears = df_Merged3.loc[:, 'NewFirstSpell_Discharge_date'].isnull()
    df_Merged3 = df_Merged3.loc[ ~Filter_NoCorrectSpellAfterTwoYears, : ]
    Filter_BeforeFirstCorrectSpell = df_Merged3.loc[:, 'Discharge_date']>=df_Merged3.loc[:, 'NewFirstSpell_Discharge_date']
    df_Merged3 = df_Merged3.loc[ Filter_BeforeFirstCorrectSpell, : ]
    
    # Time difference between admission and discharge date of first spell
    df_Merged3.loc[:, 'SpellLevel_TimeDiff'] = (df_Merged3.loc[:, 'Admission_date']-df_Merged3.loc[:, 'NewFirstSpell_Discharge_date']).dt.days
    Filter_FirstSpells = df_Merged3.loc[:, 'SpellLevel_TimeDiff']<0 # For first correct spell, we would have TimeDiff = -LOS
    df_Merged3.loc[Filter_FirstSpells, 'SpellLevel_TimeDiff']=0
    
    # New error flag after changing NewFirstSpell_Discharge_date
    Filter_Errors = (df_Merged3.loc[:, 'SpellLevel_TimeDiff']>0) & (df_Merged3.loc[:, 'SpellLevel_ICD10_'+ConditionName]==0)
    df_Merged3.loc[:, 'SpellLevel_Error_'+ConditionName] = np.where(Filter_Errors, 1, 0)
    
    return df_Merged3
# --------------------------------------------------

df_dementia_EthnicityIMD = Dementia_LinkEthnicityIMD(df_DementiaSpellLevel_ini, df_dementiaEthnicity_ini)
print(df_dementia_EthnicityIMD.columns)
df_dementia_EthnicityIMD.head()
```

```python

# Link ethinicity data for Autism dataset
# --------------------------------------------------
def Autism_LinkEthnicityIMD(df_ini, df_Ethnicity_ini):
    df = df_ini.copy()
    df_Ethnicity = df_Ethnicity_ini.copy()
    ConditionName = 'Autism'

    # Dropping unwanted index columns | Renaming columns
    ColumnsUnnamed = ['Unnamed: 0', 'Unnamed: 0.1']
    for col in ColumnsUnnamed:
        if col in df.columns: 
            df = df.drop(columns=[col])

    df_Ethnicity = df_Ethnicity.rename(columns={'HESID1':'HESID'})

    # Convert date features
    DateFeatures = ['Admission_date', 'Discharge_date', 'EPIstart', 'EPIend', 'EPIend_First']
    for datefeature in DateFeatures:
        df.loc[:, datefeature] = pd.to_datetime(df.loc[:, datefeature], format='%Y-%m-%d')
        
    # Find discharge date of first spell
    Filter_Diagnosis = df.loc[:, 'SpellLevel_ICD10_'+ConditionName]==1
    df_First = df.loc[Filter_Diagnosis, :].sort_values(by=['Admission_date', 'EPIstart', 'EPIend']).drop_duplicates(subset=['HESID'], keep='first')
    df = pd.merge(df, df_First[['HESID', 'Discharge_date']], on='HESID', how='left')
    df = df.rename(columns={'Discharge_date_x':'Discharge_date',
                            'Discharge_date_y':'SpellLevel_Discharge_date_First'})    

    # Ethnicity column
    ColumnList = ['HESID', 'Ethnicity', 'HFRS_score', 'HFRS_Band', 'IMD_score', 'IMD_decile']
    df_Ethnicity = df_Ethnicity.loc[:, ColumnList].drop_duplicates(subset=['HESID'], keep='last')
    df = df.drop(columns=['IMD Score']) # old IMD Score feature with filled-in NaNs using mean values
    df_Merged = pd.merge(df, df_Ethnicity, on='HESID', how='left')
        
    # Correct IMD scores: AFTER 2013
    df_Merged2 = df_Merged.rename(columns={'IMD_score':'IMD Score',
                                           'IMD_decile': 'IMD Decile'})

    # Exclude all spells with discharge date < cut-off-date
    Filter_AfterFirstTwoYears = (df_Merged2.loc[:, 'Discharge_date']>pd.to_datetime('2013-04-01', format='%Y-%m-%d'))
    df_Merged3 = df_Merged2.loc[Filter_AfterFirstTwoYears, :]

    # Find new date for first spell with code: NewFirst
    Filter_Code = (df_Merged3.loc[:, 'SpellLevel_ICD10_'+ConditionName]==1)
    df_NewFirst = df_Merged3.loc[Filter_Code, :].sort_values(by=['Admission_date',
                                                                 'EPIstart', 'EPIend']).drop_duplicates(subset=['HESID'], keep='first')
    ColumnList = ['HESID', 'Discharge_date', 'Admission_date']
    df_Merged3 = pd.merge(df_Merged3, df_NewFirst.loc[:, ColumnList], on='HESID', how='left').rename(columns={'Discharge_date_x':'Discharge_date',
                                                                                                              'Discharge_date_y':'NewFirstSpell_Discharge_date',
                                                                                                              'Admission_date_x':'Admission_date',
                                                                                                              'Admission_date_y':'NewFirstSpell_Admission_date'})
    # Exclude (First spell < cut-off-date) & (spell < NewFirst)
    Filter_NoCorrectSpellAfterTwoYears = df_Merged3.loc[:, 'NewFirstSpell_Discharge_date'].isnull()
    df_Merged3 = df_Merged3.loc[ ~Filter_NoCorrectSpellAfterTwoYears, : ]
    Filter_BeforeFirstCorrectSpell = df_Merged3.loc[:, 'Discharge_date']>=df_Merged3.loc[:, 'NewFirstSpell_Discharge_date']
    df_Merged3 = df_Merged3.loc[ Filter_BeforeFirstCorrectSpell, : ]
    
    # Time difference between admission and discharge date of first spell
    df_Merged3.loc[:, 'SpellLevel_TimeDiff'] = (df_Merged3.loc[:, 'Admission_date']-df_Merged3.loc[:, 'NewFirstSpell_Discharge_date']).dt.days
    Filter_FirstSpells = df_Merged3.loc[:, 'SpellLevel_TimeDiff']<0 # For first correct spell, we would have TimeDiff = -LOS
    df_Merged3.loc[Filter_FirstSpells, 'SpellLevel_TimeDiff']=0
    
    # New error flag after changing NewFirstSpell_Discharge_date
    Filter_Errors = (df_Merged3.loc[:, 'SpellLevel_TimeDiff']>0) & (df_Merged3.loc[:, 'SpellLevel_ICD10_'+ConditionName]==0)
    df_Merged3.loc[:, 'SpellLevel_Error_'+ConditionName] = np.where(Filter_Errors, 1, 0)
    
    return df_Merged3
# --------------------------------------------------

df_Autism_EthnicityIMD = Autism_LinkEthnicityIMD(df_AutismSpellLevel_ini, df_AutismEthnicity_ini)
print(df_Autism_EthnicityIMD.columns)
df_Autism_EthnicityIMD.head()
```
